```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------
# 1. Load data from SQLite
# -----------------------
db_path = r"C:\Users\MSI GF66\PycharmProjects\Conversational_AI_chatbot\assistant_v14.db"
conn = sqlite3.connect(db_path)

query = "SELECT user_id, messages, calculations, searches, code_executions FROM stats"
df = pd.read_sql_query(query, conn)

conn.close()

print("Loaded stats:")
print(df)

# -----------------------
# 2. Extract usage numbers
# -----------------------
stats = df.iloc[0]  # first row
labels = ["Messages", "Calculations", "Searches", "Code executions"]
values = [
    stats["messages"],
    stats["calculations"],
    stats["searches"],
    stats["code_executions"]
]

# -----------------------
# 3. BAR CHART
# -----------------------
plt.figure(figsize=(6, 4))
plt.bar(labels, values)
plt.title("Chatbot Usage Statistics")
plt.ylabel("Count")
plt.savefig("stats_bar_chart.png", dpi=200)
plt.show()

# -----------------------
# 4. PIE CHART
# -----------------------
plt.figure(figsize=(6, 6))
plt.pie(values, labels=labels, autopct="%1.1f%%")
plt.title("Usage Distribution")
plt.savefig("stats_pie_chart.png", dpi=200)
plt.show()

# -----------------------
# 5. HORIZONTAL BAR CHART
# -----------------------
plt.figure(figsize=(6, 4))
plt.barh(labels, values)
plt.title("Usage Overview")
plt.xlabel("Count")
plt.savefig("stats_horizontal_chart.png", dpi=200)
plt.show()

# -----------------------
# 6. Print summary
# -----------------------
print("\n===== SUMMARY =====")
for label, value in zip(labels, values):
    print(f"{label}: {value}")
```
